```python
import pandas as pd 
import numpy as np
```

```python
ebola_data = pd.read_csv('F:/Pandemic-Database/Ebola/Processed_Data/Ebola_CDC_Confirmed_and_Death_Clean_Data.csv')
ebola_data
```

### Guinea Data

```python
guinea_confirmed_data = ebola_data[(ebola_data['Sub-region'] == ' Guinea') & (ebola_data['Type'] == 'Confirmed')]
guinea_confirmed_data = guinea_confirmed_data.reset_index(drop = True)
guinea_confirmed_data
```

```python
guinea_confirmed_data['Case_diff'] = guinea_confirmed_data['Case Number'].diff()
guinea_confirmed_data = guinea_confirmed_data[guinea_confirmed_data['Case_diff'] >= 0]
guinea_confirmed_data = guinea_confirmed_data.reset_index(drop = True)
guinea_confirmed_data
```

```python
guinea_confirmed_daily_diff_data = guinea_confirmed_data[['Date','Case_diff']]
guinea_confirmed_daily_diff_data = guinea_confirmed_daily_diff_data.rename(columns={"Date":'date',
                                                                                   "Case_diff":'Guinea_Ebola_Confirmed_Case_Number'})
guinea_confirmed_daily_diff_data
```

```python
guinea_confirmed_daily_diff_data.plot()
```

```python
guinea_confirmed_daily_diff_data.to_csv('F:/Pandemic-Database/Ebola/Processed_Data/guinea_confirmed_daily_diff_data.csv',
                                       index=False)
```
