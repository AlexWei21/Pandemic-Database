```python
import pandas as pd 
import numpy as np
```

```python
afr_countries = "Algeria, Angola, Benin, Botswana, Burkina Faso, Burundi, Cameroon, Cape Verde, Central African Republic, Chad, Comoros, Ivory Coast, Democratic Republic of the Congo, Equatorial Guinea, Eritrea, Ethiopia, Gabon, Gambia, Ghana, Guinea, Guinea-Bissau, Kenya, Lesotho, Liberia, Madagascar, Malawi, Mali, Mauritania, Mauritius, Mozambique, Namibia, Niger, Nigeria, Republic of the Congo, Rwanda, Senegal, Seychelles, Sierra Leone, South Africa, South Sudan, Eswatini, Togo, Uganda, Tanzania, Zambia, Zimbabwe"
amr_countries = "Antigua and Barbuda, Argentina, Bahamas, Barbados, Belize, Bolivia, Brazil, Canada, Chile, Colombia, Costa Rica, Cuba, Dominica, Dominican Republic, Ecuador, El Salvador, Grenada, Guatemala, Guyana, Haiti, Honduras, Jamaica, Mexico, Nicaragua, Panama, Paraguay, Peru, Saint Kitts and Nevis, Saint Lucia, Saint Vincent and the Grenadines, Suriname, Trinidad and Tobago, United States, Uruguay, Venezuela"
sear_countries = "Bangladesh, Bhutan, North Korea, India, Indonesia, Maldives, Myanmar, Nepal, Sri Lanka, Thailand, Timor-Leste"
eur_countries = "Albania, Andorra, Armenia, Austria, Azerbaijan, Belarus, Belgium, Bosnia and Herzegovina, Bulgaria, Croatia, Cyprus, Czech Republic, Denmark, Estonia, Finland, France, Georgia, Germany, Greece, Hungary, Iceland, Ireland, Israel, Italy, Kazakhstan, Kyrgyzstan, Latvia, Lithuania, Luxembourg, Malta, Moldova, Monaco, Montenegro, Netherlands, North Macedonia, Norway, Poland, Portugal, Romania, Russia, San Marino, Serbia, Slovakia, Slovenia, Spain, Sweden, Switzerland, Tajikistan, Turkey, Turkmenistan, Ukraine, United Kingdom, Uzbekistan"
emr_countries = "Afghanistan, Bahrain, Djibouti, Egypt, Iran, Iraq, Jordan, Kuwait, Lebanon, Libya, Morocco, Oman, Pakistan, Qatar, Saudi Arabia, Somalia, Sudan, Syria, Tunisia, United Arab Emirates, Yemen"
wpr_countries = "Australia, Brunei, Cambodia, China, Cook Islands, Fiji, Japan, Kiribati, Laos, Malaysia, Marshall Islands, Micronesia, Mongolia, Nauru, New Zealand, Niue, Palau, Papua New Guinea, Philippines, Samoa, Singapore, Solomon Islands, South Korea, Taiwan, Tonga, Tuvalu, Vanuatu, Vietnam"

afr_countries = afr_countries.split(sep=', ')
amr_countries = amr_countries.split(sep=', ')
sear_countries = sear_countries.split(sep=', ')
eur_countries = eur_countries.split(sep=', ')
emr_countries = emr_countries.split(sep=', ')
wpr_countries = wpr_countries.split(sep=', ')
```

```python
country_region_dict = {'AFRO': afr_countries,
                      'AMRO': amr_countries,
                      'SEARO': sear_countries,
                      'EURO': eur_countries,
                      'EMRO': emr_countries,
                      'WPRO': wpr_countries}
country_region_dict
```

```python
df_afr_countries = pd.DataFrame(afr_countries, columns=['Country'])
df_afr_countries['Region'] = 'AFRO'

df_amr_countries = pd.DataFrame(amr_countries, columns=['Country'])
df_amr_countries['Region'] = 'AMRO'

df_sear_countries = pd.DataFrame(sear_countries, columns=['Country'])
df_sear_countries['Region'] = 'SEARO'

df_eur_countries = pd.DataFrame(eur_countries, columns=['Country'])
df_eur_countries['Region'] = 'EURO'

df_emr_countries = pd.DataFrame(emr_countries, columns=['Country'])
df_emr_countries['Region'] = 'EMRO'

df_wpr_countries = pd.DataFrame(wpr_countries, columns=['Country'])
df_wpr_countries['Region'] = 'WPRO'
```

```python
country_region_list = pd.concat([df_afr_countries, df_amr_countries,df_sear_countries,df_eur_countries,df_emr_countries,df_wpr_countries])
country_region_list.index = country_region_list['Country']
country_region_list = country_region_list.drop(columns=['Country'])
country_region_dict = country_region_list.to_dict()['Region']
country_region_dict
```
