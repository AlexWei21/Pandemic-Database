```python
import pandas as pd 
import numpy as np
```

```python
mpox_data = pd.read_csv('F:/Pandemic-Database/Raw_Data/Monkeypox/owid-monkeypox-data.csv')

mpox_data = mpox_data[['location','date','total_cases','total_deaths','new_cases','new_deaths']]

mpox_data.insert(0,'Microbe Family', 'Poxviridae')
mpox_data.insert(1,'Microbe Genus', 'Orthopoxvirus')
mpox_data.insert(2,'Microbe Species', 'MPXV')
mpox_data.insert(3,'Region', np.nan)
mpox_data.insert(5,'Domain', np.nan)
mpox_data.insert(6,'Sub-Domain', np.nan)

mpox_data = mpox_data.rename(columns={'location':'Country',
                                     'total_cases':'Cumulative_Cases',
                                     'total_deaths':'Cumulative_Deaths',
                                     'new_cases':'Cases',
                                     'new_deaths':'Deaths'})

mpox_data
```

```python
country_to_region_dict = np.load('F:/Pandemic-Database/Internal/Intermediate_Data_Archive/Country_Region_Dict.npy',
                                 allow_pickle=True).item()

country_to_region_dict.update({'Aruba':'AMRO',
'Bermuda':'AMRO',
'Congo':'AFRO',
'Democratic Republic of Congo':'AFRO',
"Curacao":'AMRO',
'Czechia':'EURO',
'Gibraltar':'EURO',
'Greenland':'AMRO',
'Guadeloupe':'AMRO',
'Guam':'WPRO',
'Martinique':'AMRO',
'New Caledonia':'WPRO',
'Puerto Rico':'AMRO',
'Saint Martin (French part)':'AMRO'})

np.save('F:/Pandemic-Database/Internal/Intermediate_Data_Archive/Country_Region_Dict.npy', country_to_region_dict)

Region_names = ['Africa','Asia','Europe','North America','Oceania','South America','World']

mpox_data = mpox_data[~mpox_data['Country'].isin(Region_names)]

mpox_data['Region'] = mpox_data['Country'].map(country_to_region_dict)

mpox_data
```

```python
mpox_data = mpox_data.melt(id_vars=['Microbe Family','Microbe Genus','Microbe Species','Region','Country','Domain','Sub-Domain','date'],
                          value_vars=['Cumulative_Cases','Cumulative_Deaths','Cases','Deaths'],
                          var_name='type',
                          value_name='number')

mpox_data
```

```python
case_data = mpox_data[mpox_data['type'] == 'Cases']
death_data = mpox_data[mpox_data['type'] == 'Deaths']
cumcase_data = mpox_data[mpox_data['type'] == 'Cumulative_Cases']
cumdeath_data = mpox_data[mpox_data['type'] == 'Cumulative_Deaths']
```

```python
case_data
```

```python
case_data.to_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Monkeypox/Mpox_World_Country_Daily_Cases.csv',
                index = False)
death_data.to_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Monkeypox/Mpox_World_Country_Daily_Deaths.csv',
                 index = False)
cumcase_data.to_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Monkeypox/Mpox_World_Country_Daily_CumCases.csv',
                index = False)
cumdeath_data.to_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Monkeypox/Mpox_World_Country_Daily_CumDeaths.csv',
                index = False)
```
