```python
import pandas as pd 
import numpy as np
```

```python
life_expectancy = pd.read_csv("F:/Pandemic-Database/Raw_Data/Meta_Data/life-expectancy.csv")
population = pd.read_csv("F:/Pandemic-Database/Raw_Data/Meta_Data/population-and-demography.csv")
healthcare_expenditure = pd.read_csv("F:/Pandemic-Database/Raw_Data/Meta_Data/total-healthcare-expenditure-gdp.csv")
```

```python
life_expectancy = life_expectancy.rename(columns={'Entity':'Country',
                                                  'Period life expectancy at birth - Sex: all - Age: 0':'Life_Expectancy'})
life_expectancy = life_expectancy[['Country','Year','Life_Expectancy']]
life_expectancy
```

```python
total_population = population[['Country name','Year','Population']]
total_population = total_population.rename(columns={'Country name':'Country',})
total_population
```

```python
healthcare_expenditure = healthcare_expenditure.rename(columns={'Entity':'Country',
                                                               'Indicator:Current health expenditure (CHE) as percentage of gross domestic product (GDP) (%)': 'Healthcare_Expenditure_Percentage'})
healthcare_expenditure = healthcare_expenditure[['Country','Year','Healthcare_Expenditure_Percentage']]
healthcare_expenditure
```

```python
healthcare_demographic_data_table = life_expectancy.merge(healthcare_expenditure,on=['Country','Year'],how='outer')
healthcare_demographic_data_table = healthcare_demographic_data_table.merge(total_population,on=['Country','Year'],how='outer')
healthcare_demographic_data_table
```

```python
healthcare_demographic_data_table.to_csv('F:/Pandemic-Database/Meta_Data/Country_Level_Healthcare_Demographic_Data.csv',
                                        index = False)
```
