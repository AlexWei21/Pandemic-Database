```python
us_state_level_population = pd.read_csv("/Users/alex/Documents/GitHub/Past-Pandemic-Metadata/Meta_Data/US_State_Population.csv")
us_state_level_population = us_state_level_population.rename(columns={"Name":"Domain",
                                                                      "Pop. 2020":"Population"})
us_state_level_population = us_state_level_population[us_state_level_population['Domain']!='United States']
us_state_level_population.insert(0,"Country","United States")
us_state_level_population.insert(2,"Sub-Domain",np.nan)
us_state_level_population.head()
```

```python
australia_state_level_population = pd.read_csv("/Users/alex/Documents/GitHub/Past-Pandemic-Metadata/Meta_Data/Australian Bureau of Statistics.csv")
australia_state_level_population = australia_state_level_population.rename(columns={"Population at 30 June 2023 ('000)":"Population",
                                                                                    "Unnamed: 0":"Domain"})
australia_state_level_population = australia_state_level_population[["Domain","Population"]]
australia_state_level_population.insert(0,"Country","Australia")
australia_state_level_population.insert(2,"Sub-Domain",np.nan)

australia_state_level_population['Population'] = australia_state_level_population['Population'] * 1000

australia_state_level_population = australia_state_level_population[:8]
australia_state_level_population
```

```python
canada_state_level_population = pd.read_csv("/Users/alex/Documents/GitHub/Past-Pandemic-Metadata/Meta_Data/Canada_State_Population.csv")
canada_state_level_population = canada_state_level_population.rename(columns={'name':'Domain',
                                              'pop2022':'Population'})
canada_state_level_population = canada_state_level_population[['Domain','Population']]
canada_state_level_population.insert(0,"Country",'Canada')
canada_state_level_population.insert(2,"Sub-Domain",np.nan)
canada_state_level_population
```
