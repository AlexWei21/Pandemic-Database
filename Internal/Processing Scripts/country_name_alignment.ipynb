```python
us_state_level_population = pd.read_csv("/Users/alex/Documents/GitHub/Past-Pandemic-Metadata/Meta_Data/US_State_Population.csv")
us_state_level_population = us_state_level_population.rename(columns={"Name":"Domain",
                                                                      "Pop. 2020":"Population"})
us_state_level_population = us_state_level_population[us_state_level_population['Domain']!='United States']
us_state_level_population.insert(0,"Country","United States")
us_state_level_population.insert(2,"Sub-Domain",np.nan)
us_state_level_population.head()
```
