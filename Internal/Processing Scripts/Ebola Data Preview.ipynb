```python
import pandas as pd 
import numpy as np
```

```python
ebola_data = pd.read_csv('F:/Pandemic-Database/Raw_Data/Ebola/Ebola_CDC_Confirmed_and_Death_Clean_Data.csv')
ebola_data
```

### Guinea Data

```python
guinea_confirmed_data = ebola_data[(ebola_data['Sub-region'] == ' Guinea') & (ebola_data['Type'] == 'Confirmed')]
guinea_confirmed_data = guinea_confirmed_data.reset_index(drop = True)
guinea_confirmed_data = guinea_confirmed_data.rename(columns= {'Virus Family':'Microbe Family',
                                                     'Virus Genus': 'Microbe Genus',
                                                     'Virus Species': 'Microbe Species',
                                                     'Sub-region': 'Country',
                                                     'Type': 'type',
                                                     'Date': 'date',
                                                     'Case Number': 'number'})
guinea_confirmed_data.insert(5, 'State/City', 'None')
guinea_confirmed_data['type'] = 'CumCases'
guinea_confirmed_data['Region'] = 'AFRO'
guinea_confirmed_data = guinea_confirmed_data.drop(columns='Comments')
guinea_confirmed_data
```

```python
guinea_confirmed_data.to_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Ebola/Ebola_Guinea_Weekly_CumCases.csv',
                            index = False)
```

```python
guinea_case_data = guinea_confirmed_data.copy()
guinea_case_data['Case_diff'] = guinea_case_data['number'].diff()
guinea_case_data = guinea_case_data[guinea_case_data['Case_diff'] >= 0]
guinea_case_data = guinea_case_data.reset_index(drop = True)
guinea_case_data['number'] = guinea_case_data['Case_diff']
guinea_case_data['type'] = 'Cases'
guinea_case_data['Region'] = 'AFRO'
guinea_case_data = guinea_case_data.drop(columns='Case_diff')
guinea_case_data
```

```python
guinea_case_data.to_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Ebola/Ebola_Guinea_Weekly_Cases.csv',
                        index = False)
```
