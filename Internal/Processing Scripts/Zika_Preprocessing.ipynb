```python
import pandas as pd 
import numpy as np
```

```python
country_level = pd.read_csv('F:/Pandemic-Database/Raw_Data/Zika/clean_zika_dataset.csv')

country_level.insert(5,'Domain',np.nan)
country_level.insert(6,'Sub-Domain',np.nan)
country_level = country_level.drop(columns=['Comments'])

country_level = country_level.rename(columns={'Virus Family':'Microbe Family',
                                             'Virus Genus':'Microbe Genus',
                                             'Virus Species':'Microbe Species',
                                             'Sub-region':'Country',
                                             'Date':'date',
                                             'Type':'type',
                                             'Case Number':'number'})

country_level['Region'] = 'AMRO'
country_level['type'] = 'Cases'

country_level
```

```python
country_level.to_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Zika/Zika_World_Country_Weekly_Cases.csv',
                    index = False)
```
