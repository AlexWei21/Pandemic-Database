```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
dengue_data = pd.read_csv("/Users/alex/Documents/GitHub/Past-Pandemic-Metadata/Raw_Data/Dengue_Fever/clean_dengue_data.csv")

dengue_data = dengue_data.rename(columns={'Virus Family': 'Microbe Family',
                                         'Virus Genus': 'Microbe Genus',
                                         'Virus Species': 'Microbe Species',
                                         'Sub-region': 'Domain',
                                         'Date':'date',
                                         'Type':'type',
                                         'Case Number': 'number'})

dengue_data.insert(4,'Country','None')
dengue_data.insert(6,'Sub-Domain', np.nan)


dengue_data['Region'] = 'AMRO'
dengue_data['Country'] = np.where(dengue_data['Domain']=='San Juan', 'Puerto Rico', 'Peru')
dengue_data['type'] = 'Cases'
dengue_data = dengue_data.drop(columns='Comments')

dengue_data = dengue_data.drop_duplicates(subset = ['Country','Domain','Sub-Domain','date','type'],
                                          keep='first')

dengue_data
```

```python
dengue_data_sanjuan = dengue_data[dengue_data['Domain'] == 'San Juan']
dengue_data_sanjuan
```

```python
dengue_data_iquitos = dengue_data[dengue_data['Domain'] == 'Iquitos']
dengue_data_iquitos
```

```python
dengue_data.to_csv("/Users/alex/Documents/GitHub/Past-Pandemic-Metadata/Processed_Data/Dengue_Fever/Dengue_AMRO_Country_Weekly_Cases.csv",
                   index = False)
```

```python
dengue_data_sanjuan.to_csv("F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Dengue Fever/Dengue_Sanjuan_Weekly_Cases.csv",
                          index = False)
dengue_data_iquitos.to_csv("F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Dengue Fever/Dengue_Iquitos_Weekly_Cases.csv",
                          index = False)
```

### Cumulative Data

```python
dengue_data = dengue_data.sort_values(by = ['Domain','date'])
dengue_cumulative_data = dengue_data.groupby(['Microbe Family','Microbe Genus','Microbe Species','Region','Country','Domain','date','type']).sum().groupby(level=5).cumsum().reset_index()
dengue_cumulative_data['Sub-Domain'] = np.nan
dengue_cumulative_data['type'] = 'Cumulative_Cases'
dengue_cumulative_data
```

```python
dengue_cumulative_data.to_csv("/Users/alex/Documents/GitHub/Past-Pandemic-Metadata/Processed_Time_Series_Data/Dengue_Fever/Dengue_AMRO_Country_Weekly_CumCases.csv",
                              index = False)
```
