```python
import os

directory = '/Users/alex/Desktop/Ebola 2015 Data/'

ebola_2015_data = pd.DataFrame()

for filename in os.listdir(directory):
    data_path = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(data_path):
        data = pd.read_csv(data_path,
                           header=None)
        data = data.fillna(" ")
        data.columns = data.loc[0] + data.loc[1]
        data = data[2:]

        ebola_2015_data = pd.concat([ebola_2015_data,data])        
```

```python
ebola_2015_data.columns = ['Country','Data as of','Case definition', 'Number of cases Cumulative', 'Number of cases in past 21 days', 'Number of deaths Cumulative']
ebola_2015_data.dropna(subset = ['Data as of'],
                      inplace = True)
```

```python
ebola_2015_data
```

```python
guinea_data = ebola_2015_data[ebola_2015_data['Country'] == 'Guinea']
guinea_data['Data as of'] = pd.to_datetime(guinea_data['Data as of'])
guinea_data = guinea_data.sort_values(by = ['Data as of'])

guinea_data.reset_index(inplace = True,
                       drop = True)

guinea_data
```

```python
guinea_data.to_csv('/Users/alex/Desktop/Ebola 2015 Data/Processed_data/guinea_full_data.csv')
```
