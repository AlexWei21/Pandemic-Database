```python
import pandas as pd
import numpy as np
```

## Global Domain Data

```python
covid_global_domain_cum_data = pd.read_csv('F:/Pandemic-Database/Raw_Data/Covid 19/time_series_covid19_confirmed_global.csv')

covid_global_domain_cum_data = covid_global_domain_cum_data.rename(columns={'Country/Region':'Country',
                                                                           'Province/State':'Domain',})

covid_global_domain_cum_data = covid_global_domain_cum_data.drop(columns=['Lat','Long'])

data_columns = covid_global_domain_cum_data.columns[2:]

covid_global_domain_cum_data = pd.melt(covid_global_domain_cum_data, 
                                       id_vars=['Domain','Country'],
                                       value_vars=data_columns,
                                       var_name = 'date',
                                       value_name='number')

covid_global_domain_cum_data = covid_global_domain_cum_data[['Country','Domain','date','number']]

covid_global_domain_cum_data.insert(0,'Microbe Family','Coronaviridae')
covid_global_domain_cum_data.insert(1,'Microbe Genus','Betacoronavirus')
covid_global_domain_cum_data.insert(2,'Microbe Species','SARS-CoV-2')
covid_global_domain_cum_data.insert(3,'Region',np.nan)
covid_global_domain_cum_data.insert(6,'Sub-Domain',np.nan)
covid_global_domain_cum_data.insert(8,'type','Cumulative_Cases')

covid_global_domain_cum_data
```

```python
country_to_region_dict = np.load('F:/Pandemic-Database/Internal/Intermediate_Data/Country_Region_Dict.npy',
                                 allow_pickle=True).item()

country_to_region_dict.update({'Burma':'SEARO',
'Cabo Verde':'AFRO',
'Congo (Brazzaville)':'AFRO',
'Congo (Kinshasa)':'AFRO',
"Cote d'Ivoire":'AFRO',
'Czechia':'EURO',
'Holy See':'EURO',
'Korea, North':'SEARO',
'Korea, South':'WPRO',
'Kosovo':'EURO',
'Liechtenstein':'EURO',
'Sao Tome and Principe':'AFRO',
'Taiwan*':'WPRO',
'US':'AMRO'})

covid_global_domain_cum_data['Region'] = covid_global_domain_cum_data['Country'].map(country_to_region_dict)

covid_global_domain_cum_data = covid_global_domain_cum_data.dropna(subset='Region')

covid_global_domain_cum_data['Country'] = covid_global_domain_cum_data['Country'].replace('US','United States')

covid_global_domain_cum_data
```

```python
covid_global_domain_cum_data = covid_global_domain_cum_data[~covid_global_domain_cum_data['Domain'].isna()]
covid_global_domain_cum_data['date'] = pd.to_datetime(covid_global_domain_cum_data['date'],format="%m/%d/%y")
covid_global_domain_cum_data
```

## US Domain Data

```python
us_domain_data = pd.read_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Covid 19/Covid_US_Daily_CumCases.csv')
us_domain_data = us_domain_data.rename(columns={'State/City':'Domain'})
us_domain_data.insert(7,'Sub-Domain',np.nan)

us_domain_data = us_domain_data[['Microbe Family', 'Microbe Genus', 'Microbe Species', 'Region', 'Country', 'Domain', 'Sub-Domain','date', 'type', 'number']]
us_domain_data['date'] = pd.to_datetime(us_domain_data['date'])
us_domain_data
```

## World Country Data

```python
world_country_level_data = pd.read_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Covid 19/Covid_World_Daily_CumCases.csv')
world_country_level_data.insert(6,'Sub-Domain',np.nan)
world_country_level_data = world_country_level_data.rename(columns={'State/City':'Domain'})
world_country_level_data['date'] = pd.to_datetime(world_country_level_data['date'])
world_country_level_data
```

## Merge Data

```python
covid_full_data = pd.concat([covid_global_domain_cum_data,us_domain_data,world_country_level_data]).sort_values(by=['date','Country','Domain','Sub-Domain'])

covid_full_data = covid_full_data.reset_index(drop=True)

covid_full_data
```

```python
covid_full_data.to_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Covid 19/Covid_World_Domain_Daily_CumCases.csv',
                      index = False)
```
