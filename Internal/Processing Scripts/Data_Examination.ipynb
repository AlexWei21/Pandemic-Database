### SARS

```python
sars_data = pd.read_csv("F:/Pandemic-Database/SARS/Clean_SARS_Data.csv")

sars_data_us = sars_data[(sars_data['Sub-region'] == 'United States') & (sars_data['Type'] == 'Cases')]
sars_data_us = sars_data_us[['Date','Case Number']]
sars_data_us['Case Number'] = sars_data_us['Case Number'].diff()
sars_data_us = sars_data_us.reset_index(drop = True)
sars_data_us = sars_data_us.dropna()
sars_data_us = sars_data_us[sars_data_us['Case Number'] >= 0]

sars_data_china = sars_data[(sars_data['Sub-region'] == 'China') & (sars_data['Type'] == 'Cases')]
sars_data_china = sars_data_china[['Date','Case Number']]
sars_data_china['Case Number'] = sars_data_china['Case Number'].diff()
sars_data_china = sars_data_china.reset_index(drop = True)
sars_data_china = sars_data_china.dropna()
sars_data_china = sars_data_china[sars_data_china['Case Number'] >= 0]
```

```python
sars_data_us.to_csv("F:/Pandemic-Database/Data_Collection/SARS_US.csv",
                    index = False)

sars_data_china.to_csv("F:/Pandemic-Database/Data_Collection/SARS_China.csv",
                       index = False)
```

```python
len(sars_data_us)
```

```python
len(sars_data_china)
```

```python
sars_data_us['Date'] = pd.to_datetime(sars_data_us['Date'])
sars_data_us = sars_data_us.set_index('Date')
plt.xticks(rotation=45, ha='right')
plt.plot(sars_data_us.index, sars_data_us['Case Number'])
plt.show()
```

```python
sars_data_china['Date'] = pd.to_datetime(sars_data_china['Date'])
sars_data_china = sars_data_china.set_index('Date')
plt.xticks(rotation=45, ha='right')
plt.plot(sars_data_china.index, sars_data_china['Case Number'])
plt.show()
```
