```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
SARS_data = pd.read_csv("F:\Pandemic-Database\Raw_Data\SARS/sars_2003_complete_dataset_clean.csv")
SARS_data = SARS_data.rename(columns={'Date':'date',
                                     'Cumulative number of case(s)':'Cumulative_Cases',
                                     'Number of deaths':'Deaths',
                                     'Number recovered':'Recovered'})
SARS_data = pd.melt(SARS_data, 
                    id_vars=['Country','date'],
                    value_vars=['Cumulative_Cases','Deaths','Recovered'],
                    var_name='type',
                    value_name='number')

SARS_data.insert(0,'Microbe Family','Coronaviridae')
SARS_data.insert(1,'Microbe Genus','Betacoronavirus')
SARS_data.insert(2,'Microbe Species','SARS-CoV-1')
SARS_data.insert(3,'Region','None')
SARS_data.insert(5,'Domain','None')
SARS_data.insert(6,'Sub-Domain','None')

SARS_data
```

```python
country_region_dict = np.load('F:/Pandemic-Database/Internal/Intermediate_Data_Archive/Country_Region_Dict.npy', allow_pickle=True).item()
country_region_dict['Hong Kong SAR, China'] = "SEARO"
country_region_dict['Macao SAR, China'] = "SEARO"
country_region_dict['Republic of Korea'] = "SEARO"
country_region_dict['Taiwan, China'] = "SEARO"
country_region_dict['Viet Nam'] = "SEARO"
country_region_dict['Russian Federation'] = "EURO"
country_region_dict['Republic of Ireland'] = "EURO"
```

```python
SARS_data['Region'] = SARS_data['Country'].map(country_region_dict)
SARS_data
```

```python
SARS_data_CumCases = SARS_data[SARS_data['type'] == 'Cumulative_Cases']
SARS_data_CumCases
```

```python
SARS_data_Recovered = SARS_data[SARS_data['type'] == 'Recovered']
SARS_data_Recovered
```

```python
pre_SARS_data_Cases = SARS_data_CumCases[['Country','date','number']]
pre_SARS_data_Cases.index = pre_SARS_data_Cases['date']
pre_SARS_data_Cases = pre_SARS_data_Cases.drop(columns=['date'])
pre_SARS_data_Cases = pre_SARS_data_Cases.groupby(['Country'],).diff()
pre_SARS_data_Cases = pre_SARS_data_Cases.reset_index(drop = True)

SARS_data_Cases = SARS_data_CumCases.copy()
SARS_data_Cases['number'] = pre_SARS_data_Cases
SARS_data_Cases['type'] = 'Cases'
SARS_data_Cases = SARS_data_Cases.dropna(subset=['number'])
SARS_data_Cases
```
