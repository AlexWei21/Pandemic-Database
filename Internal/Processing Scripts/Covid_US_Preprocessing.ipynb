```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
covid_us_data = pd.read_csv('F:/Pandemic-Database/Raw_Data/Covid 19/us-states.csv')

covid_us_data = covid_us_data[['date','state','cases','deaths']]
covid_us_data = covid_us_data.rename(columns={'state':'State/City',
                                              'cases':'Cumulative_Cases',
                                             'deaths':'Cumulative_Deaths'})
covid_us_data = pd.melt(covid_us_data, 
                       id_vars=['date','State/City'],
                       value_vars=['Cumulative_Cases','Cumulative_Deaths'],
                       var_name='type',
                       value_name='number')

covid_us_data
```

```python
covid_us_data.insert(0,'Microbe Family','Coronaviridae')
covid_us_data.insert(1,'Microbe Genus','Betacoronavirus')
covid_us_data.insert(2,'Microbe Species','SARS-CoV-2')
covid_us_data.insert(3,'Region','AMRO')
covid_us_data.insert(4,'Country','United States')

covid_us_data
```

```python
covid_us_data.groupby('State/City').count()
```

```python
covid_us_case_data = covid_us_data[covid_us_data['type']=='Cumulative_Cases']
covid_us_death_data = covid_us_data[covid_us_data['type']=='Cumulative_Deaths']
```

```python
covid_us_case_data.to_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Covid 19/Covid_US_Daily_CumCases.csv',
                          index = False)
covid_us_death_data.to_csv('F:/Pandemic-Database/Past_Pandemic_Time_Series_Data/Covid 19/Covid_US_Daily_CumDeaths.csv',
                          index = False)
```
